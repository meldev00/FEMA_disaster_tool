```python
import pandas as pd
```

```python
df= pd.read_csv('./datasets/29.8549,-89.9906.csv')
df1= pd.read_csv('./datasets/29.8797,-90.4340.csv')
df2= pd.read_csv('./datasets/29.9511,-90.0715.csv')
df3= pd.read_csv('./datasets/30.0463,-89.9837.csv')
```

```python
df= pd.concat([df,df1,df2,df3],ignore_index=True)
```

```python
df['name'].drop_duplicates(inplace=True)
```

```python
df
```

```python
df['Website_FEMA']= df['FEMA Lifeline'].replace([
                                                'Safety & Security',
                                                'Food, Water & Shelter',
                                                'Health & Medical',
                                                'Energy',
                                                'Communications',
                                                'Transportation'],[1,2,3,4,5,6])
```
